# Faire une regression avec l'algorithme du gradient

L'algorithme du gradient est un algorithme d'optimisation qui permet de trouver le minimum de toute fonction convexe. Assez utile d'ailleurs, car en cherchant le minimum de la loss function (une fonction convexe qui calcul l'erreur entre la valeur obtenue et la valeur de référence(donnée)) ou encore la cost function qui n'est que la moyenne des erreurs.
Chercher le minimum de la cost fonction reviendrait à touver le modèle où le taux d'erreur est le plus faible. C'est de ce fait, ce qui est recherché lors d'une regression, on cherche la droite de regression. 

```python
#ne pas hésiter à changer la valeur du learning rate ou de l'epsilon selon les données
from math import*
import pandas
import matplotlib.pyplot as plt
import numpy
 
donnees = pandas.read_csv('nuage de pts.csv')
x=donnees["X"]
y=donnees["Y"]
plt.scatter(x, y)
plt.show()

m = 0 #coef directeur
p = 0 #ordonnée à l'origine
alpha= 1e-5 #learning Rate(=le pas)
n = len(x)

i = 0
epsilon = 1e-9
f=2

while numpy.linalg.norm(f) > epsilon and i < 1000: #test d'arrêt : si le gradient est inférieur à un e (>=0) posé
  y_pred = m*x+p
  dm = (-2/n)*numpy.sum(x*(y-y_pred))
  dp = (-2/n)*numpy.sum(y - y_pred)

  m = m-dm*alpha
  p = p-dp*alpha   
  f = numpy.array([dm,dp])
  i = i + 1

print(m,p)

y_pred = m*x+p
plt.plot([min(x), max(x)], [min(y_pred), max(y_pred)], color='orange')
plt.scatter(x,y)

plt.show()
```
